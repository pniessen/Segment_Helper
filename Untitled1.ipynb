```r
library(rvest)
library(httr)

searchterms <- c(
'T-Shirts',
# 'Athletic T-Shirts',
# 'Tank tops',
# 'Sports jerseys',
# 'Polo shirts',
# 'Fleece tops',
# 'Baselayer tops',
# 'Sports bras',
# 'Athletic shorts',
# 'Athletic pants',
# 'Compression pants',
# 'Yoga pants',
# 'Fleece bottoms',
# 'Baselayer bottoms',
# 'Windbreakers',
# 'Sweatshirts',
# 'Jackets',
# 'Running shoes',
# 'Everyday sneakers',
# 'Training shoes',
# 'Basketball shoes',
# 'Baseball cleats',
# 'Football cleats',
# 'Soccer cleats',
# 'Athletic bags',
# 'Backpacks',
# 'Athletic gloves',
# 'Athletic socks',
# 'Socks',
'Hats')
```

```r
searchterms <- gsub(" ","+",searchterms)
searchterms <- gsub("-","+",searchterms)
prices <- list()
uastring <- "Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36"
```

```r
for(i in searchterms){
  searchurl <- html_session(paste0("http://www1.macys.com/shop/search?keyword=",i), user_agent(uastring))
  itemprices <- html_text(html_nodes(searchurl,".prices span"))
  prices[[i]] <- itemprices
}
```

```r
for(i in searchterms){
  prices[[i]] <- c(prices[[i]],rep(".",400 - length(prices[[i]])))
}
```

```r
write.csv(prices,file="productprices_2.csv")
```
